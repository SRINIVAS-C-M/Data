```python
import pycaret
```

```python
import numpy as np
import pandas as pd

df = pd.read_csv("/content/regression_housing_price.csv")
df.head()
```

```python
# Here median house value is our target column which is price
df.rename(columns = {'median_house_value':'price'}, inplace = True)
```

```python
data = df.sample(frac=0.9, random_state=2)
data_unseen = df.drop(data.index)

data.reset_index(drop=True, inplace=True)
data_unseen.reset_index(drop=True, inplace=True)

print('Data for Modeling: ' + str(data.shape))
print('Unseen Data For Predictions: ' + str(data_unseen.shape))
```

```python
from pycaret.regression import setup, compare_models, evaluate_model, predict_model, save_model, load_model
reg = setup(data = data, target = 'price', session_id=123, fold_shuffle=True,experiment_name = 'house_price_prediction')
```

```python
best_model = compare_models()
```

```python
evaluate_model(best_model)
```

```python
predict_model(best_model,data=data_unseen)
```

```python
save_model(best_model, 'regression')
```
